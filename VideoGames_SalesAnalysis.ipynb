```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("E:NDownloads/Video_Games_Sales_as_at_22_Dec_2016.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
sns.distplot(df.Year_of_Release.dropna(),kde = False , bins = 39)
```

```python
sns.set(style = 'whitegrid')
ax = sns.boxplot(x=df.Global_Sales , showfliers = False)
```

```python
plt.figure(figsize = (16,5))
sns.countplot(x=df.Platform.dropna(),order = df.Platform.value_counts().index)
```

```python
plt.figure(figsize = (10,4))
sns.countplot(x=df.Genre.dropna(),order = df.Genre.value_counts().index)
```

```python
plt.figure(figsize=(15,6))
corr = df.corr()
sns.heatmap(corr,xticklabels=corr.columns.values,
           yticklabels=corr.columns.values,annot=True)
```

```python
plt.scatter(df.Global_Sales.fillna('0'),df.NA_Sales.fillna('0'))
plt.title('Scatter Plot')
plt.xlabel('Global Sales')
plt.ylabel('Nortth America Sales')
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.pairplot(df)
```
